#### Ridge vs Lasso
Approaches to reduce model complexity and overfitting

- **Ridge:** extension of the linear regression, with the addition of penalty term to the loss function. The coefficients are reduced but never go to zero, which means all features are kept, but their impact is reduced.
- **Lasso:** allow coefficients to go to zero. can be helpful in feature selection and is better for problem with many useless features;

**General Guidelines:**
- Start with Lasso if you believe many features are irrelevant
- Use Ridge if you are dealing with multicollinearity or if you want to keep all features but reduce
their impact
- Consider ElasticNet for a balance between the two approaches
- Cross-validation can help choose between Ridge and Lasso

**ElasticNet:** best of both worlds, adding both L1 and L2 in the regularization. But it is more computationally heavy, as it has twice the amounts of parameters to estimate.
