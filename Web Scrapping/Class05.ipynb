```python
import pandas as pd
from joblib import Parallel, delayed
import re
from tqdm import tqdm
```

```python
df = pd.read_csv("/Users/eduardo/Downloads/all_ECB_speeches.csv", sep="|").dropna()
```

```python
def preprocess(text):
    return re.sub("[^a-z\s]", "", text.lower())

results = Parallel(2)(
    delayed(preprocess)(text) for text in tqdm(df["contents"].tolist())
)
```
